```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data_jobs.csv')
df.head()
#df.city.value_counts()
df.drop(['Unnamed: 0'],axis='columns',inplace = True)
```

```python
df['CITY']= df['city'].str.split(', ').str.get(0).str.title()# seperating city and state, create new column just for city 
df.head(10)
```

```python
df['state']= df['city'].str.split(',').str.get(-1).str.upper()#create new column for state
df['state']= df['state'].str.strip(' ').str.split(' ').str.get(0)
df.head()
```

```python
df[df['state'].str.contains('NEW')]['city'].unique()#looking for some items which start with the same name:(NEW)
```

```python
df[df['state'].str.contains('WEST')]['city'].unique()
```

```python
df['state'] = np.where( (df['state']=='NEW')& (df['city']=='New York State'),
                            'NY',
                            df['state'])
df['state'] = np.where( (df['state']=='NEW')& (df['city']=='New Jersey'),
                            'NJ',
                            df['state'])
```

```python
df= df.replace({'state': {'NORTH':'NC','CALIFORNIA':'CA','TENNESSEE':'TN','MISSOURI':'MO',
                      'IOWA':'IA','FLORIDA':'FL','WASHINGTON':'WA','KENTUCKY':'KY',
                      'TEXAS':'TX','ARIZONA':'AZ','OREGON':'OR','WEST':'WV','COLORADO':'CO',
                          'ILLINOIS':'IL','MASSACHUSETTS':'MA','MINNESOTA':'MN',
                          'ALABAMA':'AL','GEORGIA':'GA','UTAH':'UT','OHIO':'OH','VIRGINIA':'VA','NEVADA':'NV'}})#need to change some of states name
```

```python
df['state'].unique()
```

```python
df.info()
```

```python
#cleaning column of salary
df['salary'].str.lower().str.strip(' a year').str.strip(' an hour').str.strip('$').str.strip('from').str.replace('$', '').value_counts()
```

```python
import numpy as np
# df['salary'].str.contains('an hour').dropna()
df['salary_hour'] = np.where(df['salary'].str.contains('an hour'),
                            df['salary'],
                            'NAN')
df['salary_hour'] = df['salary_hour'].str.strip(' an hour').str.replace('$', '').str.strip('From').str.strip('Up to')
```

```python
df['salary_hour_min'] = df['salary_hour'].str.split(' - ').str.get(0).astype('float')
```

```python
df['salary_hour_max'] = df['salary_hour'].str.split(' - ').str.get(1).astype('float')
```

```python
df.head()
```

```python
df['salary_year'] = np.where(df['salary'].str.contains('a year'),
                            df['salary'],
                            'NAN')
```

```python
df['salary_mean']= np.where(df['salary_hour_max'].isnull()&df['salary_hour_min'].notnull()
             ,df['salary_hour_min'],df[["salary_hour_min", "salary_hour_max"]].mean(axis=1))
```

```python
df['state'].unique()
```

```python
df['salary_year'].value_counts()
```

```python
df['salary_year'] = df['salary_year'].str.strip(' a year').str.replace('$', '').str.strip('From').str.strip('Up to').str.replace(',', '')
```

```python
df['salary_year'].value_counts()
```

```python
df['salary_year_min'] = df['salary_year'].str.split(' - ').str.get(0).astype('float')
```

```python
df['salary_year_max'] = df['salary_year'].str.split(' - ').str.get(1).astype('float')
```

```python
df['salary_year_mean']= np.where(df['salary_year_max'].isnull()&df['salary_year_min'].notnull()
             ,df['salary_year_min'],df[["salary_year_min", "salary_year_max"]].mean(axis=1))
```

```python
df.head()
```

```python
df=df[['title', 'company', 'rating', 'salary', 'CITY', 'state'
        , 'salary_mean',
       'salary_year_mean']]
```

```python
df.head(50)
```

```python
df.rename(columns={'salary_mean':'salary_hour_mean'},inplace= True)
```

```python
#convert hourly salary to yearly salary and make save it in the same column name :'salary_year_mean'
df['salary_year_mean']=np.where(df['salary_hour_mean'].notnull()&df['salary_year_mean'].isnull(),
                                df['salary_hour_mean']*37.5*52,
                                df['salary_year_mean'] )
```

```python
#take a look 
df.describe()
```

```python
#deleting unusing columns like: salary_hour_mean
df=df[['title', 'company', 'rating','CITY', 'state'
        ,'salary_year_mean']]
```
