```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# creating the list of urls
path_links = ['https://www.indeed.com/q-data-scientist-l-usa-jobs.html']
stirng= 'https://www.indeed.com/jobs?q=data+scientist&l=usa&start='
for i in range(1,200):
    path_links.append(stirng+str(i*10))
```

```python
city_col = []
rating_col = []
salary_col =[]
title_col =[]
company_col=[]
n=0
for url in path_links:

    # url = "https://www.indeed.com/jobs?q=data+scientist&l=usa"
    response = requests.get(url)
    page_content = BeautifulSoup(response.content)
    job_div = page_content.find_all('div', attrs ={'class': 'jobsearch-SerpJobCard'})


    for i in range(len(job_div)):
        city_div = job_div[i].select(".accessible-contrast-color-location")
        city = city_div[0].text.strip('\n')
        city_col.append(city)
#         print(city_div[0].text)
        title_div= job_div[i].select(".jobtitle")
        title = title_div[0].text.strip('\n')
        title_col.append(title)
        company_div = job_div[i].select('.company')
        company = company_div[0].text.strip('\n')
        company_col.append(company)

        try:
            salary_div = job_div[i].select(".salaryText")
            salary = salary_div[0].text.strip('\n')
            salary_col.append(salary)
            
        except:
#             print("")
            salary_col.append('')
           
        try:
            rating_div = job_div[i].select(".ratingsContent")
            rating = rating_div[0].text.strip('\n')
            rating_col.append(rating)

        except:
#             print("")    
            rating_col.append('')
        
        n+=1
```

```python
len(city_col)
```

```python
jobs=[title_col,company_col,city_col,rating_col,
salary_col]
```

```python
df = pd.DataFrame(jobs, index = ['title','company','city','rating','salary']).transpose()
df.head()
```
